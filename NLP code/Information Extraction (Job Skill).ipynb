## information extraction

### Year of Work Experience extraction

```python
def get_year(text):

    #find the years text
    years = re.findall(r'[0-9+-]* year',text)
    years = re.findall(r'[0-9]*',str(years))
    year = max(years)
    return year
```

```python
data['year'] = data['FullDescription'].apply(get_year)
```

### skill modifier extraction

```python
# Read the skill name from the datafile
def get_skill_list():
    skill = pd.read_csv('data/Skillset.csv',header=None)
    skill.columns = ["type","name"]
    skillset = skill['name'].tolist()
    return skillset

# Dividing paragraphs into sentences
def split_text(text): 
    text = re.split('[,|.]',text)
    return text

def get_wordnet_pos(tag): # Use Parts of Speech to tag words
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return None

def get_modifier_lemmas(text):# tokenization
    tokens = word_tokenize(text)
    result = [[word,tag] for word,tag in pos_tag(tokens) if tag in ['NN','NNS','JJ','JJR','JJS','RB','RBS','RBR']]# Modifier

    wnl = WordNetLemmatizer()
    lemmas_sent = []
    for tag in result:
        wordnet_pos = get_wordnet_pos(tag[1]) or wordnet.NOUN
        lemmas_sent.append(wnl.lemmatize(tag[0], pos=wordnet_pos)) # lemmatization
    
    return lemmas_sent


def get_skill_info(jd):
    skillset = get_skill_list()
    col = {}
    
    for skill in skillset: # select Sentences with key skills
        match_list = []
        for sentence in jd:
            if ' '+skill in sentence:
                
                match_list.append(sentence)
        
        if match_list:
            words = get_modifier_lemmas(str(match_list)) # Extract the modifiers in the keyword sentences and clean the modifiers.
            words.append('mentioned')
            col[skill]= words
    return col


def jd_skill_df(skill): # Generate dataframe for jd-skill
    
    jd_skill_df = pd.DataFrame(skill,index=[0]).T
    
    return jd_skill_df
    

def get_match_info(jd_serial, skillset): # main function
    
    jd_skill = pd.DataFrame(jd_serial)
    
    jd_skill['FullDescription']=jd_skill['FullDescription'].apply(split_text)
    match_list = []
```

```python
# apply the funtion to data
jd_list = pd.DataFrame(data)
jd_list['year'] = jd_list['FullDescription'].apply(get_year)
jd_list['slipt_jd'] = jd_list['FullDescription'].apply(split_text)
jd_list['info'] = jd_list['slipt_jd'].apply(get_skill_info)
jd_list = jd_list['info'].apply(pd.Series)
jd_list
```

### Skill level labeling

```python
corpus = jd_list.values.tolist()
def flat(nums):
    res = []
    for i in nums:
        if isinstance(i, list):
            res.extend(flat(i))
        else:
            res.append(i)
    return res
corpus = flat(corpus)
```

```python
# Export all modifiers for analysis and annotation
corpus = pd.DataFrame(data=corpus)
count = pd.DataFrame(corpus.value_counts())
count.head(1000)

count.to_csv('./corpus_output.csv')
```
