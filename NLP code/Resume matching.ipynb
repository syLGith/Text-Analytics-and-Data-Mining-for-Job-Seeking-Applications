```python
import pandas as pd
from nltk.corpus import stopwords
import re

# load data
#df = pd.read_csv('./jobs_small.csv', encoding="latin-1")
df_data = pd.read_excel('./data/jd_clean.xlsx')
```

```python
for i in df_data.index:
    with open('./data/'+str(i)+'.txt', 'w',encoding='utf-8') as f:
        f.write(str(df_data.loc[i,'Title'])+ '\n')
        f.write(str(df_data.loc[i,'FullDescription']))
```

```python
 def clean_text(text):
        STOPWORDS = set(stopwords.words('english'))
        # change to lower-csae
        text = str(text).lower()
        # drop the stopwords
        text = ' '.join(word for word in text.split() if word not in STOPWORDS) 
        return text
```

```python
def consine_similarity(resume, job):
    doc = [resume,job]
    cv = CountVectorizer()
    count_matrix = cv.fit_transform(doc)
    matchpercentage = cosine_similarity(count_matrix)[0][1]
    matchpercentage = round(matchpercentage,2)

    return(matchpercentage)
```

```python
df_data['score']=''
```

```python
for i in df_data.index:
    #jd=df_data.loc[i,'FullDescription']
    f = open('./data/'+str(i)+'.txt', 'r',encoding = 'utf -8')      
    jd = f.read()
    resume = docx2txt.process('./data/resume.docx')
    resume = clean_text(resume)
    #print (i,df_data.loc[i,'Title'], consine_similarity(resume,jd))
    df_data.loc[i,'score']=consine_similarity(resume,jd)
```

```python
df_data = df_data.sort_values(by=['score'],ascending=False)
df_data.head(20)
```
